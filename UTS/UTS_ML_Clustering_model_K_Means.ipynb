Clustering adalah metode pengelompokan data ke dalam beberapa cluster atau kelompok sehingga data dalam satu cluster memiliki tingkat kemiripan yang maksimum dan data antar cluster memiliki kemiripan yang minimum. Salah satu algoritma clustering yang paling sederhana dan umum adalah K-Means Clustering. K-Means Clustering merupakan algoritma unsupervised learning yang berfungsi untuk mengelompokkan data ke dalam beberapa cluster. Tujuan dari K-Means Clustering adalah meminimalisasi object function yang telah diatur pada proses clustering dengan cara minimalisasi variasi antar 1 cluster dengan maksimalisasi variasi dengan data di cluster lainnya




Langkah-langkah dalam algoritma K-Means Clustering adalah sebagai berikut


*   Menentukan jumlah cluster
*   Secara acak mendistribusikan data cluster
*   Menghitung rata-rata dari data yang ada di cluster
*   Menggunakan langkah baris 3 kembali sesuai nilai treshold







Kelebihan dari K-Means Clustering adalah cepat dalam proses clustering. Namun, K-Means Clustering juga memiliki kekurangan, seperti sensitif terhadap nilai centroid, hasil dari K-Means selalu berubah-ubah, sulit untuk memilih jumlah cluster yang tepat, overlapping, dan kegagalan dalam konverge

promt = beri kode untuk membuat kmeans dengan jumlah cluster = 10 dan buat kolom baru bernama label dan visualisasikan berdasarkan label, buat untuk df_scorer dan df_assist

```python
#Library
from sklearn.cluster import KMeans
```

```python
# Memilih fitur yang akan digunakan
X_scorer = df_scorer[['goals', 'played', 'goals_per90', 'total_shots', 'goal_conversion', 'shot_accuracy']]
X_assist = df_assist[['assits', 'goals', 'played', 'chances_created', 'chances_per90', 'total_passes']]
```

```python
# Standarisasi fitur
scaler = StandardScaler()
X_scorer_scaled = scaler.fit_transform(X_scorer)
X_assist_scaled = scaler.fit_transform(X_assist)
```

```python
# Membuat model KMeans dengan jumlah cluster = 10
kmeans_scorer = KMeans(n_clusters=10, random_state=42)
kmeans_assist = KMeans(n_clusters=10, random_state=42)

cluster_labels_scorer = kmeans_scorer.fit_predict(X_scorer)
cluster_labels_assist = kmeans_assist.fit_predict(X_assist)
```

```python
# Menambahkan kolom label pada df_scorer dan df_assist
df_scorer['Label'] = cluster_labels_scorer
df_assist['Label'] = cluster_labels_assist
```

```python
# Melakukan clustering pada dataset df_scorer dan df_assist
kmeans_scorer.fit(X_scorer_scaled)
kmeans_assist.fit(X_assist_scaled)
```

```python
# Visualisasi hasil clustering
plt.figure(figsize=(10, 6))
```

```python
# Memvisualisasikan data untuk setiap kluster pada df_scorer
for label in range(10):
    cluster_data = df_scorer[df_scorer['label'] == label]
    plt.scatter(cluster_data['goals'], cluster_data['assists'], label=f'Cluster {label}')

# Menampilkan centroid
plt.scatter(kmeans_scorer.cluster_centers_[:, 0], kmeans_scorer.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centroid')

plt.title('Clustering Result for df_scorer')
plt.xlabel('Goals')
plt.ylabel('Assists')
plt.legend()
plt.show()

# Memvisualisasikan data untuk setiap kluster pada df_assist
for label in range(10):
    cluster_data = df_assist[df_assist['label'] == label]
    plt.scatter(cluster_data['assits'], cluster_data['chances_created'], label=f'Cluster {label}')

# Menampilkan centroid
plt.scatter(kmeans_assist.cluster_centers_[:, 0], kmeans_assist.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centroid')

plt.title('Clustering Result for df_assist')
plt.xlabel('Assists')
plt.ylabel('Chances Created')
plt.legend()
plt.show()
```

```python
# Melakukan prediksi dengan model yang sudah dilatih untuk df_scorer
new_data_scorer = { 'goal': [12, 8], 'played': [11, 8], 'goals_per90': [1.28, 1.15], 'total_shots': [35, 18], 'goal_conversion': [0.34, 0.44], 'shot_accuracy': [0.74, 0.72]}
new_df_scorer = pd.DataFrame(data=new_data_scorer)
new_pred_scorer = kmeans_scorer.predict(new_df_scorer)
print(new_pred_scorer)

# Melakukan prediksi dengan model yang sudah dilatih untuk df_assist
new_data_assist = {'played': [10, 7], 'chances_created': [16, 12], 'chances_per90': [1.6, 1.08], 'total_passes': [341, 406], 'goals': [4, 3], 'assists': [6, 2]}
new_df_assist = pd.DataFrame(data=new_data_assist)
new_pred_assist = kmeans_assist.predict(new_df_assist)
print(new_pred_assist)
```

```python
#Library
from sklearn.metrics import silhouette_score
```

```python
# Menghitung silhouette score untuk k = 2 hingga k = 10
silhouette_scores_scorer = []
silhouette_scores_assist = []

for k in range(2, 11):
    kmeans_scorer = KMeans(n_clusters=k, random_state=42)
    kmeans_assist = KMeans(n_clusters=k, random_state=42)
    cluster_labels_scorer = kmeans_scorer.fit_predict(X_scorer)
    cluster_labels_assist = kmeans_assist.fit_predict(X_assist)
    silhouette_scores_scorer.append(silhouette_score(X_scorer, cluster_labels_scorer))
    silhouette_scores_assist.append(silhouette_score(X_assist, cluster_labels_assist))

# Menampilkan nilai silhouette score untuk setiap nilai k pada df_scorer
plt.plot(range(2, 11), silhouette_scores_scorer, marker='o')
plt.title('Silhouette Score for df_scorer')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.show()

# Menampilkan nilai silhouette score untuk setiap nilai k pada df_assist
plt.plot(range(2, 11), silhouette_scores_assist, marker='o')
plt.title('Silhouette Score for df_assist')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
#Library
from sklearn.metrics import davies_bouldin_score
```

```python
# Menghitung Davies Bouldin index score untuk k = 2 hingga k = 10 pada df_scorer
davies_bouldin_scores_scorer = []
davies_bouldin_scores_assist = []

for k in range(2, 11):
    kmeans_scorer = KMeans(n_clusters=k, random_state=42)
    kmeans_assist = KMeans(n_clusters=k, random_state=42)
    cluster_labels_scorer = kmeans_scorer.fit_predict(X_scorer)
    cluster_labels_assist = kmeans_assist.fit_predict(X_assist)
    davies_bouldin_scores_scorer.append(davies_bouldin_score(X_scorer, cluster_labels_scorer))
    davies_bouldin_scores_assist.append(davies_bouldin_score(X_assist, cluster_labels_assist))

# Menampilkan nilai Davies Bouldin index score untuk setiap nilai k pada df_scorer
print('Davies Bouldin index score for df_scorer:')
for k, score in zip(range(2, 11), davies_bouldin_scores_scorer):
    print(f'k = {k}: {score}')

# Menampilkan nilai Davies Bouldin index score untuk setiap nilai k pada df_assist
print('Davies Bouldin index score for df_assist:')
for k, score in zip(range(2, 11), davies_bouldin_scores_assist):
    print(f'k = {k}: {score}')
```
