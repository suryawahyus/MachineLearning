```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')
```

```python
# Tentukan path ke file dataset di Google Drive
file_path = '/content/drive/MyDrive/Machine Learning/dataset/concrete_data.csv'

# Baca dataset menggunakan pandas
dataset = pd.read_csv(file_path)

# Tampilkan beberapa baris pertama dataset
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset_columns = dataset.columns
```

```python
pred = dataset[dataset_columns[dataset_columns!="Strength"]]
targ = dataset["Strength"]
```

```python
targ.head()
```

```python
pred_normal = ((pred - pred.mean())/pred.std())
pred_normal.head()
```

```python
n_cols = pred_normal.shape[1]
```

```python
def regression_model():
    model = Sequential()
    model.add(Dense(50,activation = "relu",input_shape = (n_cols,)))
    model.add(Dense(50,activation ="relu"))
    model.add(Dense(50,activation ="relu"))
    model.compile(optimizer = "adam", loss = "mean_squared_error")

    return model
```

```python
model = regression_model()
model.fit(pred_normal,targ,validation_split = 0.3, epochs = 50, verbose = 1)
```
